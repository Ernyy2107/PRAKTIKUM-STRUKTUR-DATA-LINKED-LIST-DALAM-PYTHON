```python
#Latihan 2.1 double linked list dengan 3 node didalamnya, kemudian lakukan penyisipan tengah dan hapus dengan kata kunci !
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node
        new_node.prev = temp

    def insert_middle(self, data):
        if self.head is None:
            return

        slow = self.head
        fast = self.head

        while fast.next and fast.next.next:
            slow = slow.next
            fast = fast.next.next

        new_node = Node(data)
        new_node.next = slow.next
        new_node.prev = slow

        if slow.next:
            slow.next.prev = new_node
        slow.next = new_node

    def delete_key(self, key):
        temp = self.head

        while temp:
            if temp.data == key:
                if temp.prev:
                    temp.prev.next = temp.next
                else:
                    self.head = temp.next
                if temp.next:
                    temp.next.prev = temp.prev
                return
            temp = temp.next

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end="\n")
            temp = temp.next
        print("None")

dll = DoubleLinkedList()

dll.append("Erny")
dll.append("Nivo")
dll.append("Martha")

print("Data awal:")
dll.display()

dll.insert_middle("Zakia")
print("\nSetelah di insert:")
dll.display()

dll.delete_key("Nivo")
print("\nSetelah hapus key 'Nivo':")
dll.display()
```

```python
#Latihan 2.2 circular linked list dengan 3 node didalamnya, kemudian lakukan penyisipan akhir dan hapus akhir !
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            return

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        temp.next = new_node
        new_node.next = self.head

    def delete_last(self):
        if self.head is None:
            return

        if self.head.next == self.head:
            self.head = None
            return
        prev = None
        temp = self.head

        while temp.next != self.head:
            prev = temp
            temp = temp.next

        prev.next = self.head

    def display(self):
        if self.head is None:
            print("List is empty")
            return

        temp = self.head
        while True:
            print(temp.data, end= "\n")
            temp = temp.next
            if temp == self.head:
                break
        print("(kembali ke head)")

cll = CircularLinkedList()

cll.append("Erny")
cll.append("Nivo")
cll.append("Martha")

print("Data awal:")
cll.display()

cll.append("Zakia")
print("\nSetelah insert diakhir:")
cll.display()

cll.delete_last()
print("\nSetelah dihapus:")
cll.display()
```
